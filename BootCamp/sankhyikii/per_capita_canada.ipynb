```python
data = pd.read_csv('canada.csv')
data.head()
```

```python
plt.scatter(data['year'],data.iloc[:,-1])
```

```python
lm = LinearRegression()
lm.fit(data[['year']],data.iloc[:,-1])
```

```python
lm.predict([[2020]])
```

```python
plt.scatter(data['year'],data.iloc[:,-1],marker='+',color='black')
plt.plot(data['year'],lm.predict(data[['year']]))
# taking a few predictions
plt.scatter([[1995,2020,2025,2010]],
            lm.predict(np.array([1995,2020,2025,2010]).reshape(-1,1)),color='red',marker='^')

for i,j in zip(lm.predict(np.array([1995,2020,2025,2010]).reshape(-1,1)),[1995,2020,2025,2010]):
    plt.annotate(int(i),(j,i))

plt.xlabel('years',fontsize=15)
plt.ylabel('prices',fontsize=15)
plt.show()
```
